# **<font color='darkorange'>Regression Tutorial</font>**

Regression is a fundamental technique in **Machine Learning** used to model relationships between variables. It is widely applied to predict outcomes based on input data.

Regression can be broadly categorized into:  
1. **<font color='blue'>Linear Regression</font>**: Assumes a linear relationship between the variables.  
2. **<font color='green'>Polynomial Regression</font>**: Captures non-linear relationships by introducing polynomial terms.

---

## **<font color='blue'>1. Linear Regression</font>**

Linear regression assumes a **straight-line relationship** between the independent variable(s) and the target variable.

---

### **<font color='blue'>Types of Linear Regression</font>**

#### **<font color='purple'>Simple Linear Regression</font>**  
A single independent variable is used to predict the target variable.  

The model predicts the output $Y$ as a function of the independent variable $X$:  

$$
Y = a + bX
$$

- **$Y$**: Predictor Variable (Dependent Variable)  
- **$X$**: Independent Variable  
- **$a$**: Intercept of the regression line (value of $Y$ when $X = 0$)  
- **$b$**: Slope of the regression line (rate of change of $Y$ when $X$ is incremented by 1 unit).  

- **<font color='darkcyan'>Practical Example</font>**:  
  A company uses **advertising budget** as the independent variable to predict **sales revenue**.  
  - **Independent Variable**: $X$ (Advertising Budget)  
  - **Dependent Variable**: $Y$ (Sales Revenue)  
  - **Goal**: Find how increasing the advertising budget impacts sales.

---

#### **<font color='purple'>Multiple Linear Regression</font>**  
Two or more independent variables are used to predict the target variable.  

The equation for Multiple Linear Regression is:  

$$
Y = a + b_1X_1 + b_2X_2 + b_3X_3 + \dots
$$

- **Coefficients**:  
  $$
  (b_1, b_2, b_3, \dots)
  $$  
- **Independent Variables**:  
  $$
  (X_1, X_2, X_3, \dots)
  $$  

- **<font color='darkcyan'>Practical Example</font>**:  
  A real estate agency uses **property features** to predict the **house price**.  
  - **Independent Variables**: $(X_1, X_2, X_3)$ (Square footage, number of bedrooms, and location rating)  
  - **Dependent Variable**: $Y$ (House Price)  
  - **Goal**: Analyze how multiple factors affect housing prices.

---

## **<font color='green'>2. Polynomial Regression</font>**

Polynomial regression extends linear regression to model **non-linear relationships** between variables. It transforms the features into polynomial terms (e.g., $X^2, X^3$).

The equation for Polynomial Regression is:  

$$
Y = a + b_1X + b_2X^2 + b_3X^3 + \dots + b_nX^n
$$

- **$Y$**: Dependent Variable  
- **$X, X^2, X^3, \dots$**: Polynomial terms  
- **Coefficients**:  
  $$
  (a, b_1, b_2, b_3, \dots)
  $$  

- **<font color='darkcyan'>Practical Example</font>**:  
   Predicting the **growth of a plant** based on **time**.  
   Initially, the plant grows slowly, then faster, and later the growth levels off, forming a curve.  
   - **Independent Variable**: $X$ (Time in days)  
   - **Dependent Variable**: $Y$ (Plant Height in cm)  
   - **Goal**: Use a polynomial regression model to predict the non-linear growth pattern over time.

---

## **<font color='darkorange'>Comparison of Linear and Polynomial Regression</font>**

| **<font color='blue'>Aspect</font>**     | **<font color='blue'>Linear Regression</font>**         | **<font color='green'>Polynomial Regression</font>**    |
|------------------------|--------------------------------|------------------------------------|
| **Nature of Relationship** | Assumes a straight-line relationship   | Models non-linear relationships (curved) |
| **Complexity**         | Simple and interpretable                 | More complex due to polynomial terms     |
| **Use Cases**          | When data follows a linear trend         | When data shows curves or non-linear trends |

---

- **<font color='blue'>Linear Regression</font>** is suitable for data with a linear relationship, like predicting sales based on advertising.  
- **<font color='green'>Polynomial Regression</font>** is ideal when the relationship is non-linear, such as growth patterns over time.  

Understanding the type of relationship in the data helps choose the correct regression technique to make accurate predictions. ðŸš€


# **<font color='darkorange'>Derivation of the Simple Linear Regression Formula</font>**

**Simple Linear Regression (SLR)** aims to model the relationship between one independent variable $X$ and one dependent variable $Y$. The relationship is represented as:

$$
Y = a + bX
$$

Where:  
- **$Y$**: Dependent variable (target)  
- **$X$**: Independent variable (predictor)  
- **$a$**: Intercept (value of $Y$ when $X = 0$)  
- **$b$**: Slope (rate of change of $Y$ with respect to $X$)  

The goal is to determine the values of the slope $b$ and intercept $a$ that minimize the **sum of squared residuals (errors)**. Letâ€™s derive the formula step by step.

---

## **<font color='blue'>1. Sum of Squared Residuals</font>**

The **residual (error)** for each data point is the difference between the observed value $Y_i$ and the predicted value $\hat{Y}_i$:

$$
e_i = Y_i - \hat{Y}_i
$$

The predicted value $\hat{Y}_i$ is given as:

$$
\hat{Y}_i = a + bX_i
$$

Thus, the residual becomes:

$$
e_i = Y_i - (a + bX_i)
$$

The objective is to minimize the **Sum of Squared Errors (SSE)**:

$$
SSE = \sum_{i=1}^n e_i^2 = \sum_{i=1}^n \left(Y_i - (a + bX_i)\right)^2
$$

---

## **<font color='blue'>2. Deriving the Slope (\(b\))</font>**

<img src="https://drive.google.com/uc?id=1pwbc0GsPHmGqScgF2WkOsuCH6eS0lGWX" alt="Regression Example" width="500"/>

To minimize **SSE**, we differentiate the **SSE** with respect to $a$ and $b$, and set the derivatives to zero.

### **Step 1: Partial Derivative with Respect to $a$**

Taking the derivative of SSE with respect to $a$:

$$
\frac{\partial SSE}{\partial a} = -2 \sum_{i=1}^n \left(Y_i - a - bX_i\right)
$$

Set this derivative to zero:

$$
\sum_{i=1}^n (Y_i - a - bX_i) = 0
$$

Rearranging, we get:

$$
\sum Y_i = na + b \sum X_i
$$

---

### **Step 2: Partial Derivative with Respect to $b$**

Taking the derivative of SSE with respect to $b$:

$$
\frac{\partial SSE}{\partial b} = -2 \sum_{i=1}^n X_i \left(Y_i - a - bX_i\right)
$$

Set this derivative to zero:

$$
\sum_{i=1}^n X_i (Y_i - a - bX_i) = 0
$$

Expanding and rearranging:

$$
\sum X_iY_i = a \sum X_i + b \sum X_i^2
$$

---

## **<font color='blue'>3. Solving for $a$ and $b$</font>**

We now have two equations:

1. $ \sum Y_i = na + b \sum X_i $  
2. $ \sum X_iY_i = a \sum X_i + b \sum X_i^2 $  

### **Solving for $b$ (Slope)**

First, substitute $a$ from the first equation into the second equation.

After simplification, the slope $b$ is given as:

$$
b = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{\sum (X_i - \bar{X})^2}
$$

Where:  
- $\bar{X}$: Mean of $X$ values  
- $\bar{Y}$: Mean of $Y$ values  

The numerator represents the **covariance** between $X$ and $Y$, and the denominator represents the **variance** of $X$.

---

### **Solving for $a$ (Intercept)**

Once we have $b$, the intercept $a$ can be calculated using:

$$
a = \bar{Y} - b\bar{X}
$$

---

## **<font color='blue'>4. Final SLR Formula</font>**

The final Simple Linear Regression equation is:

$$
Y = a + bX
$$

Where:  
- **$b$ (Slope)**:  

$$
b = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{\sum (X_i - \bar{X})^2}
$$  

- **$a$ (Intercept)**:  

$$
a = \bar{Y} - b\bar{X}
$$

---

## **<font color='darkorange'>Summary</font>**

1. **Slope $b$** measures the rate of change of $Y$ with respect to $X$ (calculated as **covariance** of $X$ and $Y$ divided by the **variance** of $X$).  
2. **Intercept $a$** determines the value of $Y$ when $X = 0$.  
3. Together, $a$ and $b$ define the best-fit line that minimizes the sum of squared residuals.

The derived formulas allow us to find the optimal regression line for any given set of data. ðŸš€


## **<font color='darkorange'>Example of Simple Linear Regression</font>**

Let's consider a real-world example of predicting a student's **exam score** based on the number of hours they study.

---

### **<font color='blue'>Problem Statement</font>**

We want to model the relationship between:  
- **<font color='green'>Independent Variable</font>** ($X$): Hours of study  
- **<font color='green'>Dependent Variable</font>** ($Y$): Exam score  

The goal is to find the equation of the best-fit line:

$$
Y = a + bX
$$

---

### **<font color='blue'>Data</font>**

| **Hours of Study ($X$)** | **Exam Score ($Y$)** |
|--------------------------|----------------------|
| 1                        | 50                   |
| 2                        | 55                   |
| 3                        | 65                   |
| 4                        | 70                   |
| 5                        | 80                   |

---

### **<font color='blue'>Step 1: Calculate the Mean</font>**

1. **Mean of $X$ (Hours of Study):**

$$
\bar{X} = \frac{\sum X_i}{n} = \frac{1 + 2 + 3 + 4 + 5}{5} = 3
$$

2. **Mean of $Y$ (Exam Score):**

$$
\bar{Y} = \frac{\sum Y_i}{n} = \frac{50 + 55 + 65 + 70 + 80}{5} = 64
$$

---

### **<font color='blue'>Step 2: Calculate the Slope ($b$)</font>**

The formula for slope $b$ is:

$$
b = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{\sum (X_i - \bar{X})^2}
$$

| **$X_i$** | **$Y_i$** | **$X_i - \bar{X}$**  | **$Y_i - \bar{Y}$** | **($X_i - \bar{X}$)($Y_i - \bar{Y}$)** | **($X_i - \bar{X}$)$^2$** |
|-----------|-----------|---------------------|---------------------|---------------------------------------|--------------------------|
| 1         | 50        | -2                  | -14                 | 28                                    | 4                        |
| 2         | 55        | -1                  | -9                  | 9                                     | 1                        |
| 3         | 65        | 0                   | 1                   | 0                                     | 0                        |
| 4         | 70        | 1                   | 6                   | 6                                     | 1                        |
| 5         | 80        | 2                   | 16                  | 32                                    | 4                        |

1. **Sum of $(X_i - \bar{X})(Y_i - \bar{Y})$:**

$$
\sum (X_i - \bar{X})(Y_i - \bar{Y}) = 28 + 9 + 0 + 6 + 32 = 75
$$

2. **Sum of $(X_i - \bar{X})^2$:**

$$
\sum (X_i - \bar{X})^2 = 4 + 1 + 0 + 1 + 4 = 10
$$

Thus, the slope $b$ is:

$$
b = \frac{75}{10} = 7.5
$$

---

### **<font color='blue'>Step 3: Calculate the Intercept ($a$)</font>**

The formula for the intercept $a$ is:

$$
a = \bar{Y} - b\bar{X}
$$

Substitute the values:

$$
a = 64 - (7.5 \times 3) = 64 - 22.5 = 41.5
$$

---

### **<font color='blue'>Step 4: Final Equation</font>**

The equation of the regression line is:

$$
Y = 41.5 + 7.5X
$$

---

### **<font color='blue'>Interpretation</font>**

- **<font color='green'>Slope ($b = 7.5$)</font>**: For every additional hour of study, the exam score increases by 7.5 points.  
- **<font color='green'>Intercept ($a = 41.5$)</font>**: If a student studies for 0 hours, their predicted exam score is 41.5.

---

### **<font color='blue'>Prediction Example</font>**

If a student studies for 6 hours ($X = 6$), the predicted score is:

$$
Y = 41.5 + 7.5(6) = 41.5 + 45 = 86.5
$$

Thus, the predicted exam score is **86.5**.

---

## **<font color='darkorange'>Summary</font>**

1. The relationship between study hours and exam scores was modeled using **Simple Linear Regression**.  
2. The final regression equation is:

$$
Y = 41.5 + 7.5X
$$


# Implementation of SLR with formula

```python
# Input Data: Hours of Study (X) and Exam Scores (Y)
X = [1, 2, 3, 4, 5]  # Independent Variable
Y = [50, 55, 65, 70, 80]  # Dependent Variable

# Step 1: Calculate Mean of X and Y
n = len(X)
mean_X = sum(X) / n
mean_Y = sum(Y) / n

# Step 2: Calculate Slope (b) and Intercept (a)
numerator = sum((X[i] - mean_X) * (Y[i] - mean_Y) for i in range(n))
denominator = sum((X[i] - mean_X) ** 2 for i in range(n))

b = numerator / denominator  # Slope
a = mean_Y - b * mean_X      # Intercept

# Step 3: Display the Results
print(f"Mean of X: {mean_X}")
print(f"Mean of Y: {mean_Y}")
print(f"Slope (b): {b}")
print(f"Intercept (a): {a}")

# Final Regression Line Equation
print(f"Regression Line: Y = {a:.2f} + {b:.2f}X")

# Step 4: Prediction
def predict(x):
    return a + b * x

# Predict Y for a new value of X
X_new = 6
Y_pred = predict(X_new)
print(f"Predicted Y for X = {X_new}: {Y_pred:.2f}")
```

# Implementation of SLR with sklearn

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Input Data: Hours of Study (X) and Exam Scores (Y)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent Variable (2D array for sklearn)
Y = np.array([50, 55, 65, 70, 80])  # Dependent Variable

# Step 1: Initialize and Train the Model
model = LinearRegression()
model.fit(X, Y)

# Step 2: Extract Slope (b) and Intercept (a)
a = model.intercept_  # Intercept
b = model.coef_[0]    # Slope

# Step 3: Display Results
print(f"Slope (b): {b}")
print(f"Intercept (a): {a}")
print(f"Regression Line: Y = {a:.2f} + {b:.2f}X")

# Step 4: Make Predictions
X_new = np.array([[6]])  # New value for prediction
Y_pred = model.predict(X_new)

print(f"Predicted Y for X = 6: {Y_pred[0]:.2f}")
```

## Plotting

```python
# Step 5: Plot the Data Points and Regression Line
import matplotlib.pyplot as plt

# Scatter plot of the data points
plt.scatter(X, Y, color='blue', label='Data Points')

# Plot the regression line
Y_line = model.predict(X)  # Predict Y values for all X values
plt.plot(X, Y_line, color='red', label='Regression Line')

# Plot the new prediction point
plt.scatter(X_new, Y_pred, color='green', s=100, marker='o', label=f'Prediction (X=6, Y={Y_pred[0]:.2f})')

# Styling the plot
plt.title("Simple Linear Regression: Study Hours vs Exam Scores")
plt.xlabel("Hours of Study")
plt.ylabel("Exam Score")
plt.legend()
plt.grid(True)
plt.show()
```

# Polynomial Regession

## **Input Data with Polynomial Features**

| Hours of Study (\(X\)) | Exam Scores (\(Y\)) | \(X^2\) (Squared Hours) |
|-------------------------|--------------------|------------------------|
| 1                       | 50                 | 1                      |
| 2                       | 55                 | 4                      |
| 3                       | 65                 | 9                      |
| 4                       | 70                 | 16                     |
| 5                       | 85                 | 25                     |


```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Input Data (Single Feature for Simplicity)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Hours of Study
Y = np.array([50, 55, 65, 70, 85])           # Exam Scores

# Step 2: Manually Create Polynomial Features (X^2)
X_poly = np.hstack((X, X**2))  # Add X^2 as a new column

# Step 3: Train the Linear Regression Model
model = LinearRegression()
model.fit(X_poly, Y)

# Step 4: Generate Predictions for Smooth Curve
X_range = np.linspace(min(X), max(X), 100).reshape(-1, 1)  # Smooth range of X values
X_range_poly = np.hstack((X_range, X_range**2))            # Add X^2 to the smooth range
Y_range_pred = model.predict(X_range_poly)

# Step 5: Plot the Results
plt.scatter(X, Y, color='blue', label='Data Points')       # Original data points
plt.plot(X_range, Y_range_pred, color='red', label='Polynomial Regression Curve')  # Regression curve

# Add labels and title
plt.title("Polynomial Regression: Hours of Study vs Exam Scores")
plt.xlabel("Hours of Study")
plt.ylabel("Exam Score")
plt.legend()
plt.grid(True)
plt.show()
```

# **<font color='darkorange'>Evaluation of Regression Fit</font>**

Evaluating the performance of a regression model is critical to understanding how well it predicts the dependent variable. This tutorial explains the key metrics, their formulas, and the advantages/disadvantages of each.

---

## **<font color='blue'>1. Key Metrics for Regression Evaluation</font>**

### **<font color='green'>1.1 Mean Absolute Error (MAE)</font>**

**Definition**: The average absolute difference between predicted values ($\hat{Y}_i$) and actual values ($Y_i$):

$$
MAE = \frac{\sum_{i=1}^n |Y_i - \hat{Y}_i|}{n}
$$

- **<font color='purple'>Interpretation</font>**: Smaller MAE indicates better model performance.  
- **<font color='purple'>Properties</font>**: Sensitive to individual errors but not to large outliers.

---

### **<font color='green'>1.2 Mean Squared Error (MSE)</font>**

**Definition**: The average of the squared differences between predicted and actual values:

$$
MSE = \frac{\sum_{i=1}^n (Y_i - \hat{Y}_i)^2}{n}
$$

- **<font color='purple'>Interpretation</font>**:  
   - MSE penalizes larger errors more than smaller ones because of squaring.  
   - Lower MSE indicates a better model.

---

### **<font color='green'>1.3 Root Mean Squared Error (RMSE)</font>**

**Definition**: The square root of the mean squared error:

$$
RMSE = \sqrt{\frac{\sum_{i=1}^n (Y_i - \hat{Y}_i)^2}{n}}
$$

- **<font color='purple'>Interpretation</font>**:  
   - RMSE is in the same units as the target variable, making it easier to interpret.  
   - Penalizes large errors more than MAE.

---

### **<font color='green'>1.4 R-squared (Coefficient of Determination)</font>**

**Definition**: R-squared explains the proportion of variance in the dependent variable that can be explained by the independent variables:

$$
R^2 = 1 - \frac{\sum_{i=1}^n (Y_i - \hat{Y}_i)^2}{\sum_{i=1}^n (Y_i - \bar{Y})^2}
$$

- **<font color='purple'>Interpretation</font>**:  
   - \(R^2\) ranges from 0 to 1.  
   - \(R^2 = 1\): Perfect fit.  
   - \(R^2 = 0\): Model explains none of the variance.

---

### **<font color='green'>1.5 Adjusted R-squared</font>**

**Definition**: Adjusted R-squared modifies the \(R^2\) value to account for the number of predictors in the model:

$$
\text{Adjusted } R^2 = 1 - \left( \frac{(1 - R^2)(n - 1)}{n - k - 1} \right)
$$

Where:  
- \(n\): Number of observations  
- \(k\): Number of predictors  

- **<font color='purple'>Interpretation</font>**:  
   - Penalizes the addition of unnecessary predictors.  
   - Useful for comparing models with different numbers of predictors.

---

## **<font color='darkorange'>Comparison of Metrics</font>**

| **<font color='blue'>Metric</font>**       | **<font color='blue'>Advantages</font>**                                          | **<font color='blue'>Disadvantages</font>**                                   |
|-----------------------|-------------------------------------------------------|-----------------------------------------------------------|
| **<font color='green'>MAE</font>**         | Simple to interpret; less sensitive to outliers.         | May not penalize large errors enough.                     |
| **<font color='green'>MSE</font>**         | Penalizes large errors more heavily.                     | Difficult to interpret due to squared units.              |
| **<font color='green'>RMSE</font>**        | Same units as target variable; penalizes large errors.    | More complex to calculate; sensitive to outliers.         |
| **<font color='green'>R-squared</font>**   | Indicates goodness of fit; easy to interpret.            | Does not penalize overfitting or account for predictors.  |
| **<font color='green'>Adjusted R-squared</font>** | Accounts for model complexity and overfitting.           | Slightly more complex to calculate.                      |

---

Using the appropriate metric depends on the specific regression task and the nature of your dataset. Combining multiple metrics often provides the most robust evaluation of your model's performance. ðŸš€


# **<font color='darkorange'>Model Selection Using Evaluation of Regression Fit</font>**

When training and evaluating multiple regression models (e.g., Linear, Polynomial, or others), we use **regression metrics** to determine which model performs the best.

---

## **<font color='blue'>1. Workflow for Model Selection</font>**

1. **<font color='green'>Train Different Models</font>**:  
   - Example: Linear Regression, Polynomial Regression (degree 2 or 3), etc.
2. **<font color='green'>Evaluate Models</font>**:  
   - Use metrics like **Mean Absolute Error (MAE)**, **Mean Squared Error (MSE)**, **Root Mean Squared Error (RMSE)**, and **R-squared ($R^2$)**.
3. **<font color='green'>Compare Performance</font>**:  
   - Select the model with the lowest error and highest $R^2$.
4. **<font color='green'>Choose the Best Model</font>**:  
   - Balance between accuracy and simplicity to avoid overfitting.

---

## **<font color='blue'>2. Example: Compare Linear vs Polynomial Regression</font>**

We have the following dataset:

| **Hours of Study ($X$)** | **Exam Scores ($Y$)** |
|---------------------------|-----------------------|
| 1                         | 50                   |
| 2                         | 55                   |
| 3                         | 65                   |
| 4                         | 70                   |
| 5                         | 85                   |

We will compare **Linear Regression** and **Polynomial Regression (degree 2)** using evaluation metrics.

---

```python

# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Input Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Hours of Study
Y = np.array([50, 55, 65, 70, 85])           # Exam Scores

# Linear Regression
model_linear = LinearRegression()
model_linear.fit(X, Y)
Y_pred_linear = model_linear.predict(X)

# Polynomial Regression (Manually create X^2)
X_poly = np.hstack((X, X**2))
model_poly = LinearRegression()
model_poly.fit(X_poly, Y)
Y_pred_poly = model_poly.predict(X_poly)

# Evaluation Metrics
def evaluate_model(Y_true, Y_pred, model_name):
    mae = mean_absolute_error(Y_true, Y_pred)
    mse = mean_squared_error(Y_true, Y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(Y_true, Y_pred)
    print(f"--- {model_name} ---")
    print(f"MAE: {mae:.2f}")
    print(f"MSE: {mse:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"R-squared: {r2:.2f}\n")

# Evaluate both models
evaluate_model(Y, Y_pred_linear, "Linear Regression")
evaluate_model(Y, Y_pred_poly, "Polynomial Regression (Degree 2)")

# Visualization
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, Y_pred_linear, color='red', label='Linear Regression')
plt.plot(X, Y_pred_poly, color='green', label='Polynomial Regression (Degree 2)')
plt.title("Model Comparison: Linear vs Polynomial Regression")
plt.xlabel("Hours of Study")
plt.ylabel("Exam Score")
plt.legend()
plt.grid(True)
plt.show()
```

---

### **<font color='green'>Step 1: Train the Models</font>**

- **Model 1**: Linear Regression  
- **Model 2**: Polynomial Regression (degree 2)

---

### **<font color='green'>Step 2: Evaluate Metrics</font>**

The metrics for both models are as follows:

| **Metric**           | **Linear Regression** | **Polynomial Regression (Degree 2)** |
|-----------------------|-----------------------|---------------------------------------|
| **MAE**              | 2.00                  | 1.20                                  |
| **MSE**              | 5.50                  | 2.29                                  |
| **RMSE**             | 2.35                  | 1.51                                  |
| **$R^2$**            | 0.96                  | 0.98                                  |

---

### **<font color='green'>Step 3: Compare Performance</font>**

- **Polynomial Regression (degree 2)** achieves lower errors and a higher $R^2$ value, indicating better fit.  
- **Linear Regression** is simpler but fails to capture the curvature in the data.

---

### **<font color='green'>Step 4: Choose the Best Model</font>**

- **Polynomial Regression (degree 2)** is the better model based on the metrics, as it provides:  
  - Lower Mean Absolute Error (MAE) and Root Mean Squared Error (RMSE).  
  - Higher $R^2$, indicating it explains more variance in the data.  

However, it is essential to validate the model on unseen data to ensure it does not overfit.

---

## **<font color='darkorange'>Key Insights</font>**

1. Use metrics to compare models and balance accuracy with complexity.  
2. Higher degree models often fit the training data better but risk overfitting.  
3. Evaluate using multiple metrics and validate with unseen data to ensure generalization.
