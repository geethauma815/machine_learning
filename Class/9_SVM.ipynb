# üéØ <span style="color:blue;">**Support Vector Machines (SVM)**</span>

Support Vector Machines (**SVMs**) are **supervised learning models** used for **classification and regression analysis**. They aim to **find the best decision boundary (hyperplane)** that separates different classes in a dataset.

---

## üìå <span style="color:green;">**Introduction to SVM**</span>
### <span style="color:purple;">**What does SVM do?**</span>
- Given a dataset with two or more classes, **SVM finds the best possible boundary** that separates the classes.
- It chooses the **optimal hyperplane** that maximizes the **margin** between different classes.

## üìå <span style="color:green;">**Understanding the Hyperplane**</span>
A **hyperplane** is a **decision boundary** that separates different classes in an **n-dimensional space**.

- **For 2D data** ‚Üí The hyperplane is a **line**.
- **For 3D data** ‚Üí The hyperplane is a **plane**.
- **For higher dimensions** ‚Üí It becomes a **higher-dimensional surface**.

SVM **chooses the hyperplane that maximizes the distance (margin) from the nearest data points** on either side.

# üìå <span style="color:green;">**Mathematics Behind SVM**</span>

Support Vector Machines (**SVMs**) are based on strong **mathematical foundations** in **linear algebra** and **optimization**. This tutorial breaks down the key **mathematical concepts** that make SVMs work.

---

## üìå <span style="color:purple;">**Understanding the Dot Product and Why It‚Äôs Used in SVM**</span>

### ‚úÖ **1. What is the Dot Product?**
The **dot product** (or **inner product**) is a fundamental operation in linear algebra that measures the **similarity between two vectors**.

For two vectors **$ a $** and **$ b $**, the dot product is:

$$
a \cdot b = |a| |b| \cos \theta
$$

where:
- **$ |a| $** = length (magnitude) of vector **$ a $**.
- **$ |b| $** = length (magnitude) of vector **$ b $**.
- **$ \theta $** = angle between the vectors.

<img src="https://drive.google.com/uc?id=1_kSgTuxpU0CPcT-sgOUKujICUHozULhr"/>

---

### ‚úÖ **2. Why Does SVM Use the Dot Product?**
The **dot product** is essential in SVM because:

- It helps **determine how aligned two vectors are**.
- It is used to **define the hyperplane** (decision boundary).
- It allows us to **compute the distance of a point from the hyperplane**.

In SVM, we use the **dot product between the weight vector $ w $ and feature vector $ x $** to classify data points.

---

## üìå <span style="color:purple;">**Defining the Hyperplane in SVM**</span>

### ‚úÖ **1. What is a Hyperplane?**
A **hyperplane** is a **decision boundary** that separates different classes.  

- In **2D**, a hyperplane is a **line**.
- In **3D**, a hyperplane is a **plane**.
- In **higher dimensions**, it is a **hyper-surface**.

### ‚úÖ **2. Mathematical Representation of the Hyperplane**
The equation of a hyperplane is:

$$
w \cdot x - b = 0
$$

where:
- **$ w $** = weight vector (determines the orientation of the hyperplane).
- **$ x $** = feature vector.
- **$ b $** = bias term (controls the offset of the hyperplane).

A point **$ x $** is classified based on its position relative to the hyperplane:

- **For Class +1**:
  $$
  w \cdot x_i - b > 0
  $$

- **For Class -1**:
  $$
  w \cdot x_i - b < 0
  $$

- **If a point lies exactly on the hyperplane**:
  $$
  w \cdot x_i - b = 0
  $$

Thus, **SVM makes predictions by computing the dot product of each point with the weight vector**.

---

## üìå <span style="color:purple;">**Maximizing the Margin in SVM**</span>

<img src="https://drive.google.com/uc?id=1s7fBOO81CgSh9WmcWIHgXkXjnB-JwlT6" width=500/>

### ‚úÖ **1. What is the Margin?**
The **margin** is the **distance between the hyperplane and the closest data points (support vectors)**.

- **SVM maximizes this margin** for **better classification**.
- A **larger margin** leads to **better generalization**.

### ‚úÖ **2. Mathematical Definition of Margin**
Using the **distance formula from a point to a hyperplane**, the distance of any point **$ x_0 $** from the hyperplane is:

$$
d = \frac{| w \cdot x_0 - b |}{\|w\|}
$$

For **support vectors**, the **closest data points to the hyperplane**, we set:

- **For Class +1**:
  $$
  w \cdot x_+ - b = 1
  $$
  Distance:
  $$
  d_+ = \frac{|1|}{\|w\|} = \frac{1}{\|w\|}
  $$

- **For Class -1**:
  $$
  w \cdot x_- - b = -1
  $$
  Distance:
  $$
  d_- = \frac{| -1 |}{\|w\|}
  $$

Thus, the **total margin** is:

$$
\text{Margin} = d_+ + d_-
$$

Substituting the values:

$$
\text{Margin} = \frac{1}{\|w\|} + \frac{1}{\|w\|}
$$

$$
\text{Margin} = \frac{2}{\|w\|}
$$

---

## üìå <span style="color:purple;">**Optimization Problem in SVM**</span>

To **ensure correct classification**, SVM enforces the following condition:

$$
y_i (w \cdot x_i - b) \geq 1
$$

where:
- **$ y_i \in \{-1,1\} $** is the class label.
- **$ x_i $** is the feature vector.
- **$ w \cdot x_i - b $** is the decision function.

### ‚úÖ **Optimization Goal: Maximizing the Margin**
To **maximize the margin**, we need to **minimize** $ \|w\| $, which leads to the **convex optimization problem**:

$$
\min \frac{1}{2} \|w\|^2
$$

subject to:

$$
y_i (w \cdot x_i - b) \geq 1, \quad \forall i
$$

This optimization problem can be solved using **Lagrange multipliers**, leading to the **dual form of SVM**.

---

## üéØ <span style="color:blue;">**Final Takeaways**</span>

üöÄ **Key Learnings from SVM Mathematics:**
- **SVM finds the best hyperplane** by maximizing the margin.
- **Dot product** measures similarity and determines classification.
- **Hyperplane equation**:  
  $$
  w \cdot x - b = 0
  $$
- **Margin formula**:  
  $$
  \frac{2}{\|w\|}
  $$
- **Optimization problem**:  
  $$
  \min \frac{1}{2} \|w\|^2, \quad y_i (w \cdot x_i - b) \geq 1
  $$

üí° **Why does SVM work so well?**
- **Maximizing margin** ‚Üí Better generalization.
- **Uses only Support Vectors** ‚Üí Less sensitive to noise.

# üìå <span style="color:green;">**Soft-Margin SVM & Nonlinear Decision Boundaries**</span>

In many real-world scenarios, **data is not perfectly linearly separable**. Support Vector Machines (**SVMs**) provide solutions by introducing **soft margins** and the **kernel trick** to handle **nonlinear decision boundaries**.

---

## üöÄ <span style="color:purple;">**Soft-Margin SVM**</span>

### ‚úÖ <span style="color:blue;">**Why Do We Need a Soft Margin?**</span>

- In **real-world datasets**, classes may **overlap** or contain **misclassified points**.
- A **hard-margin SVM** assumes **perfect linear separability**, which is **not always possible**.
- **Soft-margin SVM** allows **some misclassification** by introducing **slack variables $ \xi_i $**.

<img src="https://drive.google.com/uc?id=18430suvmDR0OOtMq5GuQQqiwEMBuZedG" width=500/>

---

### ‚úÖ <span style="color:blue;">**Mathematical Formulation**</span>

To handle misclassified points, we introduce **slack variables $ \xi_i \geq 0 $** for each data point **$ x_i $**, modifying the SVM constraint to:

$$
y_i ( w \cdot x_i + b ) \geq 1 - \xi_i, \quad \forall i
$$

where:
- **$ \xi_i = 0 $** ‚Üí correctly classified point.
- **$ 0 < \xi_i < 1 $** ‚Üí point inside the margin but on the correct side.
- **$ \xi_i \geq 1 $** ‚Üí misclassified point.

The new **optimization objective** becomes:

$$
\min \frac{1}{2} \|w\|^2 + C \sum_{i=1}^{n} \xi_i
$$

where **$ C $** is a hyperparameter that **controls the trade-off** between **maximizing the margin** and **minimizing misclassifications**.

---

## üöÄ <span style="color:purple;">**Nonlinear Decision Boundaries & The Kernel Trick**</span>

### ‚úÖ <span style="color:blue;">**Why Do We Need Kernels?**</span>

- Some datasets **cannot be separated** by a simple **straight line (hyperplane)**.
- Instead of using **a linear separator**, we **map data to a higher-dimensional space** where a **hyperplane can separate them**.

<img src="https://drive.google.com/uc?id=1qfFA5TlflzgZ3EM_THRQqSnfos2KYvaA"/>

---

### ‚úÖ <span style="color:blue;">**Mathematical Concept of Kernels**</span>

- We transform data **$ x_i $** from the original input space **to a higher-dimensional feature space** using a mapping function:

$$
\phi: x \rightarrow \phi(x)
$$

- The new decision boundary is found **in this transformed space**.

However, computing **explicit transformations** for **high-dimensional spaces** is computationally expensive. Instead, we use **Kernel functions** that compute:

$$
K(x_i, x_j) = \phi(x_i) \cdot \phi(x_j)
$$

This allows us to **implicitly map data to higher dimensions** **without directly computing $\phi(x)$**.

---

## üöÄ <span style="color:purple;">**Numerical Example: How Kernels Make Data Linearly Separable**</span>

Consider the **following 1D dataset**:

| $x$ | Class |
|----|------|
| -2 | +1  |
| -1 | +1  |
|  0 | -1  |
|  1 | -1  |
|  2 | +1  |

- In **1D**, we cannot separate this data using a single threshold.
- We **transform** the data using:

$$
\phi(x) = x^2
$$

After transformation:

| $x$ | $x^2$ | Class |
|----|----|------|
| -2 | 4  | +1  |
| -1 | 1  | +1  |
|  0 | 0  | -1  |
|  1 | 1  | -1  |
|  2 | 4  | +1  |

Now, in **Transformed space** (where the feature is $ x^2 $), a **linear hyperplane** (a straight line) can separate the data!

```python
import numpy as np
import matplotlib.pyplot as plt

# Original 1D dataset
x = np.array([-2, -1, 0, 1, 2])
y = np.array([1, 1, -1, -1, 1])  # Class labels

# Transforming data to higher dimension using phi(x) = x^2
x_transformed = x**2

# Create figure for plotting
plt.figure(figsize=(12, 5))

# üîπ Subplot 1: Original 1D Data (Non-Linearly Separable)
plt.subplot(1, 2, 1)
plt.scatter(x[y == 1], np.zeros(sum(y == 1)), color='blue', label='Class +1', s=100)
plt.scatter(x[y == -1], np.zeros(sum(y == -1)), color='red', label='Class -1', s=100)
plt.axhline(0, color='black', linewidth=0.5)
plt.xlabel("Feature x")
plt.title("Original 1D Data (Not Linearly Separable)")
plt.legend()

# üîπ Subplot 2: Transformed 2D Data (Linearly Separable)
plt.subplot(1, 2, 2)
plt.scatter(x_transformed[y == 1], np.zeros(sum(y == 1)), color='blue', label='Class +1', s=100)
plt.scatter(x_transformed[y == -1], np.zeros(sum(y == -1)), color='red', label='Class -1', s=100)
plt.axhline(0, color='black', linewidth=0.5)
plt.xlabel("Transformed Feature x¬≤")
plt.title("Transformed Data (Linearly Separable)")
plt.legend()

# Display the plots
plt.tight_layout()
plt.show()
```

---

## üöÄ <span style="color:purple;">**Common Kernel Functions**</span>

### ‚úÖ <span style="color:blue;">**1. Linear Kernel**</span>
Used when data is **already linearly separable**.

$$
K(x_i, x_j) = x_i \cdot x_j
$$

### ‚úÖ <span style="color:blue;">**2. Polynomial Kernel**</span>
Used when data **has curved decision boundaries**.

$$
K(x_i, x_j) = (x_i \cdot x_j + c)^d
$$

where:
- **$ c $** is a constant.
- **$ d $** is the polynomial degree.

### ‚úÖ <span style="color:blue;">**3. Radial Basis Function (RBF) Kernel**</span>
Used when **data is highly complex and nonlinearly distributed**.

$$
K(x_i, x_j) = \exp\left(-\frac{\|x_i - x_j\|^2}{2\sigma^2}\right)
$$

where:
- **$ \sigma $** controls the **spread of the decision boundary**.

## üéØ <span style="color:green;">**Final Takeaways**</span>

‚úÖ **Soft-margin SVM** allows misclassification by introducing **slack variables**.  
‚úÖ **The kernel trick** enables **SVM to classify nonlinearly separable data** without explicitly transforming it.  
‚úÖ **Common kernels** include **Linear, Polynomial, and RBF Kernels**.  
‚úÖ **Choosing the right kernel** depends on the **nature of the dataset**.  



```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, ConfusionMatrixDisplay

# Creating the dataset
data = {
    'Height': [6.00, 5.92, 5.58, 5.92, 5.00, 5.50, 5.42, 5.75],
    'Weight': [180, 190, 170, 165, 100, 150, 130, 150],
    'Foot_Size': [12, 11, 12, 10, 6, 8, 7, 9],
    'Gender': ['Male', 'Male', 'Male', 'Male', 'Female', 'Female', 'Female', 'Female']
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Encode target variable (Male=1, Female=0)
label_encoder = LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])

# Extract features and target
X = df[['Height', 'Weight', 'Foot_Size']]
y = df['Gender']

# Standardize features for SVM
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train SVM model (no train-test split, trained on all data)
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale')
svm_model.fit(X_scaled, y)

# Predict on the same data
y_pred = svm_model.predict(X_scaled)

# Display classification report
print("Classification Report:\n", classification_report(y, y_pred, target_names=label_encoder.classes_))

# Plot confusion matrix
ConfusionMatrixDisplay.from_estimator(svm_model, X_scaled, y, display_labels=label_encoder.classes_)
plt.show()
```
